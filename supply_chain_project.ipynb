```python
final_df.to_csv('Resultant Table.csv')
```

# Output Data 2

```python
final_df['Result'] = np.where(final_df['Bill Difference'] == 0, 'Correctly Charged',
                             np.where(final_df['Bill Difference']>0, 'Under Charged', 
                                     np.where(final_df['Bill Difference']<0, 'Over Charged', None)))
```

```python
final_df['Result'].unique()
```

```python
final_summary = final_df.groupby('Result').agg({'Order ID':'count', 'Billing Amount (Rs.)':'sum'})
```
